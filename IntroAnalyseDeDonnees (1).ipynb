```python
ibm.shape
```

```python
ibm.describe()
```

```python
sb.set_theme()
sb.catplot(
    data=ibm,
    x='JobSatisfaction', y='Attrition')
```

```python
sb.set_theme()
sb.boxplot(
    data=ibm,
    x='JobSatisfaction', y='Attrition')
```

```python
#question4 : On peut en conclure que il n'y a pas de valeur abérrantes pour ce boxplot
#question5 : Il n'y en a pas 
ibm.describe()
```

```python
sb.set_theme()
sb.barplot(
    data=ibm,
    x='TotalWorkingYears', y='Attrition')
```

```python
sb.set_theme()
sb.catplot(
    data=ibm,
    x='TotalWorkingYears', y='Attrition', kind='box', aspect=2)
```

```python
#Sur ce boxplot, on a bien des valeurs aberrantes qui se dessinent.
#on va donc les isoler
tmp2=ibm[ ibm['TotalWorkingYears']>20]
tmp2
```

```python
tmp2.Attrition=tmp2['Attrition'].replace([0,1],['No','Yes'])
sb.catplot(data=tmp2, x='TotalWorkingYears', y='Attrition', kind='box')
```

```python
ibm.describe()
```

```python
ibm['Attrition'] = ibm['Attrition'].replace(['Yes','No'],[1,0])
ibm.head(10)
```

```python
sb.catplot(data=ibm, x='EducationField', y='Attrition', kind='bar', height=10, aspect=1 )
```
