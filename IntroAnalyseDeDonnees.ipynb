```python
import pandas as pd
import numpy as np
import seaborn as sb
```

# Introduction à l'analyse de données
## Exercice 1

```python
hp=pd.read_csv('./house_pricing.csv')
hp.head(10)
```

```python
hp['MSSubClass'], hp['LotFrontage'], hp['LotArea']
```

```python
sb.set_theme()
sb.relplot(
    data=hp,
    x="SalePrice", y="LotArea"
)
```

```python
sb.set_theme()
sb.relplot(
    data=hp[ (hp['SalePrice'] < 500000) & (hp['LotArea'] < 20000)],
    x="SalePrice", y="LotArea"
)
```

```python
sb.set_theme()
sb.relplot(
    data=hp,
    x='LotFrontage', y='LotArea')
```

```python
sb.set_theme()
sb.relplot(
    data=hp[ (hp['LotFrontage'] < 200) & (hp['LotArea'] < 100000)],
    x='LotFrontage', y='LotArea')
```
